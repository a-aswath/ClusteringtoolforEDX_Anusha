```python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output
from sklearn.cluster import MiniBatchKMeans
from mpl_toolkits.axes_grid1 import make_axes_locatable

# Dummy cluster data
cluster_hierarchy = {}

# Dropdown to select the cluster to sub-cluster
cluster_dropdown = widgets.Dropdown(
    options=[i for i in range(10)],
    description='Cluster:',
    disabled=False,
)

# Textbox to input the number of sub-clusters
subcluster_input = widgets.IntText(
    value=2,
    description='Sub-clusters:',
    disabled=False
)

# Button to perform sub-clustering
button = widgets.Button(
    description='Sub-cluster'
)

# Button to go back to the previous level
back_button = widgets.Button(
    description='Back',
    disabled=True  # Initially disabled because there's no previous level yet
)

# Radio buttons to select the sub-cluster for further sub-clustering
subcluster_radio = widgets.RadioButtons(
    options=[],
    description='Sub-cluster:',
    disabled=True
)

# Output widget to display images and interaction
output = widgets.Output()

# Track the current level of clustering to facilitate recursive sub-clustering
current_level = {}

def display_selected_cluster(change):
    """Display the image of the selected cluster."""
    cluster = cluster_dropdown.value
    with output:
        clear_output(wait=True)
        fig, axes = plt.subplots(3, 5, figsize=(25, 15))  # Create a 3x5 grid for 15 tiles

        # Loop over the 15 images to display in the 3x5 grid
        for i in range(3):
            for j in range(5):
                tile_cnt = i * 5 + j  # Tile index for this subplot
                ax = axes[i, j]
                try:
                    ax.imshow((labels == cluster).reshape(15, 1024, 1024)[tile_cnt], cmap='gray')
                except:
                    ax.imshow(np.zeros((1024, 1024)), cmap='gray')  # Handle missing images
                ax.set_title(f'Cluster {cluster} Tile {tile_cnt}')
                ax.axis('off')  # Optional: Turn off axis labels

        plt.tight_layout()  # Adjust subplots to fit into figure area
        plt.show()

    # Reset level tracking and disable the back button
    current_level[cluster] = 0
    back_button.disabled = True
    subcluster_radio.options = []
    subcluster_radio.disabled = True

def on_button_click(b):
    """Perform sub-clustering and update the sub-cluster radio buttons."""
    global cluster_hierarchy, current_level

    cluster = cluster_dropdown.value
    sub_clusters = subcluster_input.value

    # Determine the current level for the selected cluster
    level = current_level.get(cluster, 0)

    # Perform sub-clustering
    mbkmeans = MiniBatchKMeans(n_clusters=sub_clusters, batch_size=200000, random_state=1)

    # Ensure that X_partial and mask have matching shapes
    image_cluster = np.zeros(X_partial_filtered.shape[0])  # Set the array shape to match X_partial

    if level == 0:
        mask = (labels == cluster)
    else:
        mask = cluster_hierarchy[cluster][level - 1][selected_subcluster]

    # Flatten the mask if necessary to match the shape of X_partial
    mask_flat = mask.ravel()  # Ensure mask is 1D to match X_partial

    # # Check if the mask and X_partial shapes match
    # if mask_flat.shape[0] != X_partial[mask_flat].shape[0]:
    #     raise ValueError("Shape mismatch between mask and X_partial.")

    # Now fit and predict the sub-clusters on the correct data slice
    mbkmeans.fit(X_partial_filtered[mask_flat])
    image_cluster[mask_flat] = mbkmeans.predict(X_partial_filtered[mask_flat])

    # Store the sub-cluster information in the hierarchy
    if cluster not in cluster_hierarchy:
        cluster_hierarchy[cluster] = {}
    cluster_hierarchy[cluster][level] = {}

    for i in range(sub_clusters):
        unique_mask = mask_flat & (image_cluster == i)
        cluster_hierarchy[cluster][level][i] = unique_mask
        print(f'Sub-cluster {i} mask sum:', np.sum(unique_mask))

    # Update radio button options for sub-clustering
    subcluster_radio.options = [f'Sub-cluster {i}' for i in range(sub_clusters)]
    subcluster_radio.disabled = False

    # Update the current level and enable the back button
    current_level[cluster] = level + 1
    back_button.disabled = False

    # Display the sub-cluster images
    with output:
        clear_output(wait=True)

        num_images = 15  # Total number of images to display
        rows, cols = 3, 5  # Create a 3x5 grid for 15 images
        tile_cnt = 0

        for j in range(sub_clusters):
            fig, ax = plt.subplots(rows, cols, figsize=(25, 15))  # 3x5 grid for 15 images

            for i in range(rows):
                for k in range(cols):
                    try:
                        im_map = ax[i][k].imshow(cluster_hierarchy[cluster][level][j].reshape(15, 1024, 1024)[tile_cnt % 15], cmap='gray')
                    except:
                        im_map = ax[i][k].imshow(np.zeros((1024, 1024)), cmap='gray')

                    ax[i][k].set_title(f'Sub-cluster {j}, Tile {tile_cnt:02d} of Cluster {cluster} (Level {level})')
                    ax[i][k].axes.get_xaxis().set_visible(False)
                    ax[i][k].axes.get_yaxis().set_visible(False)

                    divider = make_axes_locatable(ax[i][k])
                    cax = divider.append_axes('right', size='5%', pad=0.05)
                    fig.colorbar(im_map, cax=cax, orientation='vertical')

                    tile_cnt += 1

                    if tile_cnt >= num_images:
                        break
                if tile_cnt >= num_images:
                    break

            plt.margins(0, 0)
            plt.axis('off')
            plt.tight_layout()
            plt.show()

def on_back_button_click(b):
    """Go back to the previous level of sub-clustering and display 15 images."""
    selected_cluster = cluster_dropdown.value
    current_level[selected_cluster] -= 1

    if current_level[selected_cluster] == 0:
        back_button.disabled = True  # Disable back button if at the root level

    level = current_level[selected_cluster] - 1  # Set to the previous level
    if level < 0:
        return  # If no previous level exists, do nothing

    # Update radio button options for the previous level's sub-clusters
    subcluster_radio.options = [f'Sub-cluster {i}' for i in cluster_hierarchy[selected_cluster][level].keys()]
    subcluster_radio.disabled = False

    # Display the sub-cluster images for the previous level
    with output:
        clear_output(wait=True)

        num_images = 15  # Total number of images to display
        rows, cols = 3, 5  # Create a 3x5 grid for 15 images
        tile_cnt = 0

        # Loop through sub-clusters for the previous level
        for j in cluster_hierarchy[selected_cluster][level].keys():
            fig, ax = plt.subplots(rows, cols, figsize=(25, 15))  # 3x5 grid for 15 images

            for i in range(rows):
                for k in range(cols):
                    try:
                        # Display the sub-cluster images
                        im_map = ax[i][k].imshow(cluster_hierarchy[selected_cluster][level][j].reshape(15, 1024, 1024)[tile_cnt % 15], cmap='gray')
                    except:
                        im_map = ax[i][k].imshow(np.zeros((1024, 1024)), cmap='gray')  # Handle missing data

                    ax[i][k].set_title(f'Sub-cluster {j}, Tile {tile_cnt:02d} of Cluster {selected_cluster} (Level {level})')
                    ax[i][k].axes.get_xaxis().set_visible(False)
                    ax[i][k].axes.get_yaxis().set_visible(False)

                    # Add colorbar to each tile
                    divider = make_axes_locatable(ax[i][k])
                    cax = divider.append_axes('right', size='5%', pad=0.05)
                    fig.colorbar(im_map, cax=cax, orientation='vertical')

                    tile_cnt += 1
                    if tile_cnt >= num_images:
                        break
                if tile_cnt >= num_images:
                    break

            plt.margins(0, 0)
            plt.axis('off')
            plt.tight_layout()
            plt.show()

def on_subcluster_select(change):
    """Display the image of the selected sub-cluster for further sub-clustering."""
    global selected_subcluster

    selected_cluster = cluster_dropdown.value
    selected_subcluster = int(subcluster_radio.value.split()[-1])

    with output:
        clear_output(wait=True)
        level = current_level.get(selected_cluster, 0) - 1  # Previous level

        fig, ax = plt.subplots(3, 5, figsize=(25, 15))  # Create a 3x5 grid for 15 images

        tile_cnt = 0
        for i in range(3):
            for j in range(5):
                try:
                    ax[i, j].imshow(cluster_hierarchy[selected_cluster][level][selected_subcluster].reshape(15, 1024, 1024)[tile_cnt], cmap='gray')
                except:
                    ax[i, j].imshow(np.zeros((1024, 1024)), cmap='gray')  # Handle missing images
                ax[i, j].set_title(f'Sub-cluster {selected_subcluster}, Tile {tile_cnt} of Cluster {selected_cluster}')
                ax[i, j].axis('off')

                tile_cnt += 1
                if tile_cnt >= 15:
                    break
            if tile_cnt >= 15:
                break

        plt.tight_layout()
        plt.show()

# Attach handlers
cluster_dropdown.observe(display_selected_cluster, names='value')
button.on_click(on_button_click)
back_button.on_click(on_back_button_click)
subcluster_radio.observe(on_subcluster_select, names='value')

# Display the widgets
display(cluster_dropdown, subcluster_input, button, back_button, subcluster_radio, output)
```
